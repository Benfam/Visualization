Visualisation of dataset

````python
column_x = 'Age'
column_y = 'Income'

#Scatter plot`
plt.figure(figsize =(10, 6))
plt.scatter(balance[column_x], balance[column_y], alpha= 0.7, edgecolors= 'w', s =80)
#Set plot title and labels
plt.title('Scatter Plot For Age vs Income')
plt.xlabel(column_x)
plt.ylabel(column_y)
#show the plot
plt.show


````

````python
x_column = 'Gender'
y_column = 'Income'

# create a bar chart
plt.figure(figsize =(6, 5))
plt.bar(balance[x_column], balance[y_column])
#Set plot title and labels
plt.title('Bar ChartFor Gender vs Income')
plt.xlabel(x_column)
plt.ylabel(y_column)
#show the plot
plt.show
````

````python
#Pie chart to display the ethnicity
ethn_counts = balance['Ethnicity'].value_counts()
ethn_counts.head()

plt.figure(figsize =(8,6))
plt.pie(ethn_counts, labels = ethn_counts.index, autopct='%1.1f%%')
#Add title
plt.title('Distribution of Ethnicity')
#Show the plot
plt.show


plt.show
````

````python
#Pie chart to display gender count
gender_counts = balance['Gender'].value_counts()
gender_counts.head()


plt.figure(figsize =(8,6))
plt.pie(gender_counts, labels = gender_counts.index, autopct='%1.1f%%')

#Add title
plt.title('Distribution of Gender')

#Show the plot
````
